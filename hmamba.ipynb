```python
# 📦 Mamba-SSM (CP312) 离线安装脚本
# 适配: Python 3.12, CUDA 12.1 (Kaggle T4/P100 默认环境)
# 这一步绝对安全，不会触发重启

import sys
import subprocess
import glob

print("--- 正在安装 Mamba-SSM (Python 3.12专用版) ---")

# 1. 设置数据集路径
# 你的数据集挂载路径
DATASET_DIR = "/kaggle/input/mamba-ssm-2-2-2-cp312-cp312-linux-x86-64"

# 2. 自动查找 whl 文件
causal_whls = glob.glob(f"{DATASET_DIR}/causal_conv1d*.whl")
mamba_whls = glob.glob(f"{DATASET_DIR}/mamba_ssm*.whl")

if not causal_whls or not mamba_whls:
    print(f"❌ 错误: 在 {DATASET_DIR} 未找到 whl 文件！")
    print("请确保你已经在右侧 Add Data 添加了 'mamba-ssm-2-2-2-cp312-cp312-linux-x86-64' 数据集。")
else:
    print(f"📦 找到 causal_conv1d: {causal_whls[0]}")
    print(f"📦 找到 mamba_ssm: {mamba_whls[0]}")
    
    # 3. 安装 (使用 --no-deps 确保不检查 Torch 版本)
    try:
        # 顺序很重要：先装 causal_conv1d
        subprocess.check_call([sys.executable, "-m", "pip", "install", causal_whls[0], "--no-deps", "--force-reinstall", "--quiet"])
        
        # 再装 mamba_ssm
        subprocess.check_call([sys.executable, "-m", "pip", "install", mamba_whls[0], "--no-deps", "--force-reinstall", "--quiet"])
        
        print("✅ 安装成功！(无需重启)")
        
        # 验证导入
        import mamba_ssm
        print(f"🚀 成功加载: mamba_ssm {mamba_ssm.__version__}")
        
    except Exception as e:
        print(f"❌ 安装过程出错: {e}")
```
